## 부모생년월일

```python
data_df = data_df[data_df["거주지역"] != "없음"]
```

```python
data_df[['연령대']].astype('str')
```

```python
data_df[['연령대']].info()
```

```python
data_df['연령대'] = pd.to_datetime(data_df['연령대'])
```

```python

def age_list(x):
    if (x.year <= 2003) & (x.year >= 1994):
        return 0
    if (x.year <= 1993) & (x.year >= 1984):
        return 1
    if (x.year <= 1983) & (x.year >= 1974):
        return 2
    if (x.year <= 1973) & (x.year >= 1964):
        return 3
    if (x.year <= 1963) & (x.year >= 1954):
        return 4
    if (x.year <= 1953) & (x.year >= 1944):
        return 5
    if (x.year <= 1943) & (x.year >= 1934):
        return 6
    else:
        return 7
    
data_df['연령대'] = data_df['연령대'].apply(age_list)
        
```
